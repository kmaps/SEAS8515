## Installing Libraries, Importing Packages, Python Scripts

In Python, importing a package allows you to use functions, classes, and variables defined in that package within your code. A package in Python is essentially a directory containing multiple modules and an `__init__.py` file (which can be empty). Modules are individual files containing Python code. By importing a package, you can leverage pre-written code and libraries to perform tasks more efficiently.

### **Tutorial: How to Import a Package in Python**

1. **Install the Package**
   Before importing a package, ensure it is installed. You can install packages using a package manager like `pip`.

   ```bash
   pip install package_name
   ```

   Replace `package_name` with the name of the package you want to install.

2. **Importing a Package**
   After installing a package, you can import it into your Python script or interactive session. Here are different ways to import packages:

   - **Import the Entire Package**

     ```python
     import package_name
     ```

     Example:
     ```python
     import math
     print(math.sqrt(16))  # Outputs: 4.0
     ```

   - **Import Specific Modules from a Package**

     ```python
     from package_name import module_name
     ```

     Example:
     ```python
     from math import sqrt
     print(sqrt(16))  # Outputs: 4.0
     ```

   - **Import Specific Functions or Classes from a Module**

     ```python
     from package_name.module_name import function_name
     ```

     Example:
     ```python
     from math import sqrt, pow
     print(sqrt(16))  # Outputs: 4.0
     print(pow(2, 3))  # Outputs: 8.0
     ```

   - **Import a Module with an Alias**

     ```python
     import package_name as alias
     ```

     Example:
     ```python
     import numpy as np
     arr = np.array([1, 2, 3])
     print(arr)  # Outputs: [1 2 3]
     ```

3. **Exploring the Package**
   Once imported, you can explore the package’s functionality using Python’s built-in `dir()` function or checking the package’s documentation.

   ```python
   import package_name
   print(dir(package_name))  # Lists attributes and methods
   ```

### Polars

#### Data Ingestion with Polars

- What it is: A modern DataFrame library for Python and Rust, similar to pandas, but designed for speed and scalability.

- Why it’s fast: Built on Rust with efficient memory use, lazy evaluation (only computes results when needed), and automatic parallelization.

- Syntax: Very similar to pandas — easy to learn if you already know pandas — but with some extra functional-style expressions (pl.col("age") > 3).

- Big data friendly: Can handle datasets much larger than RAM using memory-mapped files and streaming.

- Lazy vs. Eager mode:

Eager: runs commands immediately (like pandas).

Lazy: builds a query plan and optimizes execution (like SQL or Spark).

- Interoperable: Can convert to/from pandas, Arrow, and NumPy easily.

- Use case: If pandas feels slow for large datasets, Polars often gives a huge speedup without much code change.

Documentation: https://docs.pola.rs/

### Numpy

- **Purpose**: `NumPy` (Numerical Python) is used for numerical computations. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

- **Key Features**:
  - **Arrays**: `NumPy` arrays are more efficient and flexible compared to Python's built-in lists. They allow for vectorized operations and broadcasting.
  - **Mathematical Functions**: It includes a range of mathematical operations and functions such as trigonometric functions, linear algebra routines, and statistical operations.
  - **Random Number Generation**: Provides functions to generate random numbers and perform random sampling.

#### Pandas and NumPy Integration
- **Pandas and NumPy**: Pandas is built on top of NumPy and uses NumPy arrays internally. You often use them together for data analysis tasks, with Pandas handling data manipulation and NumPy providing numerical operations.

In summary, `pandas` is great for handling and analyzing structured data, while `NumPy` excels in numerical computations and array operations. They are both essential tools for data scientists and analysts working with data in Python.

Documentation: https://numpy.org/doc/

````python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Perform mathematical operations
print("Add 20 to each element: ", arr + 10, '\n')        # Add 10 to each element
print("Calculate the mean of the numpy array: ", np.mean(arr), '\n')    # Calculate the mean
print("Calculate the square root of each element: ", '\n', np.sqrt(arr), '\n')    # Calculate the square root of each element

# Create a 2D array
matrix = np.array([[1, 2], [3, 4]])

# Matrix operations
print("Performing matrix multiplication: ", '\n', np.dot(matrix, matrix))  # Matrix multiplication
````

Numpy is particularly useful for its linear algebra functionality.

````python
A = np.array([[1, 2],
              [3, 4]])

# Compute the inverse of the matrix
A_inv = np.linalg.inv(A)
print("The inverse of the array is: ", A_inv, '\n')

# Compute the determinant of the matrix
det_A = np.linalg.det(A)
print("The determinant of the array is: ", det_A, '\n')

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
````

#### Eager vs Lazy Data Loading

Eager (Normal) Reading

```python
df = pl.read_parquet("data/sample.parquet")
```

#### What happens
- The file is read right away
- Data is fully loaded into memory
- Operations execute immediately
- Simple and intuitive, but less memory-efficient

#### Mental model
> Read everything first.

---

### Lazy Reading

```python
lazy_df = pl.scan_parquet("data/sample.parquet")
```

#### What happens
- No data is read initially
- A query plan (computation graph) is created
- Operations are recorded, not executed
- Execution occurs only when `.collect()` is called

#### Mental model
> Plan first, then read only what you need.

---

### Key Difference
> Eager execution loads data immediately; lazy execution builds a plan and reads only the necessary data.

#### Data Augmentation with Albumentations

Albumentations is a fast image augmentation library designed for ML and CV workflows.

*   Very fast (NumPy + OpenCV backend)
*   Clean, readable API
*   Works with images, masks, boxes, keypoints
*   Widely used in ML pipelines

Documentation: https://albumentations.ai/docs/

````python
!pip install albumentations
````

````python
import albumentations as A
import matplotlib.pyplot as plt
````

````python
transform = A.Compose([
    A.HorizontalFlip(p=0.9),
    A.RandomBrightnessContrast(p=0.3),
    A.Rotate(limit=30, p=0.7),
    A.GaussianBlur(blur_limit=3, p=0.2)
])
````

````python
augmented = transform(image=img)
aug_img = augmented["image"]

plt.imshow(aug_img)
plt.axis("off")
````

A really useful component of albumentations is its built in capability to also augment the labels (bounding boxes, segmentations, etc) so that separate logic does not have to be written.

````python
img_mask = cv2.imread("Datasets/dogs2_mask.jpg")
img_mask = cv2.resize(img_mask, (810, 540))
````

````python
img_mask.shape
````

````python
img.shape
````

````python
transform = A.Compose(
    [
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=15, p=0.5)
    ],
    additional_targets={"mask": "mask"}
)

augmented = transform(image=img, mask=img_mask)
aug_img = augmented["image"]
aug_mask = augmented["mask"]
````

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Original image
axes[0, 0].imshow(img)
axes[0, 0].set_title("Original Image")
axes[0, 0].axis("off")

# Original mask
axes[0, 1].imshow(img_mask, cmap="gray")
axes[0, 1].set_title("Original Mask")
axes[0, 1].axis("off")

# Augmented image
axes[1, 0].imshow(aug_img)
axes[1, 0].set_title("Augmented Image")
axes[1, 0].axis("off")

# Augmented mask
axes[1, 1].imshow(aug_mask, cmap="gray")
axes[1, 1].set_title("Augmented Mask")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
````

Altair is a declarative visualization library based on the Grammar of Graphics.  

Instead of manually controlling plot elements, you describe *what* you want to visualize, and Altair handles *how* it is rendered.

````python
import altair as alt
from vega_datasets import data

cars = data.cars()
alt.Chart(cars).mark_point().encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    color="Origin",
    tooltip=["Name","Origin","Horsepower","Miles_per_Gallon"]
)
````
